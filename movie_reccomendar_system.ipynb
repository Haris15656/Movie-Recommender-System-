```python
#merge both data on title
df1=credits
df2=movies
df1.columns = ['id','tittle','cast','crew']
df = df2.merge(df1,on='id')
df
```

```python

c = df['vote_average'].mean()
print (c)
```

```python
m = df['vote_count'].quantile(0.95)
qualified_movies = df.copy().loc[df['vote_count']>=m]
```

```python
def weighted_rating(x):
    v = x['vote_count']
    r = x['vote_average']
    return (v/(v+m) *r) + (m/(v+m) *c)
```

```python
qualified_movies['score'] = qualified_movies.apply(weighted_rating,axis=1)
```

```python
#Sort movies based on score calculated above
qualified_movies = qualified_movies.sort_values('score', ascending=False)

#Print the top 10 movies
qualified_movies[['title','vote_count','vote_average','score']].head(10)
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
#merge both data on title
data=movies.merge(credits,on='title')
```

```python
data.head()
```

```python
 #selecting the columns we want
 data=data[['movie_id','title','overview','genres','keywords','cast','crew']]
 data.head()
```

```python
data.iloc[0].crew
```

```python
data.isnull().sum()
data.dropna(inplace=True)
data.duplicated().sum()
```

```python
#now we have to make a data with movieid,title and tags, tags will have a paragraphy joining overview,genres,keywords,cast and crew
import ast
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l

def convert2(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      l.append(i['name'])
    else: break
  return l

def convert3(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
       l.append(i['name'])
       break
  return l
```

```python
data['genres']=data['genres'].apply(convert)
data['keywords']=data['keywords'].apply(convert)
data['cast']=data['cast'].apply(convert2)
data['crew']=data['crew'].apply(convert3)

```

```python
data
```

```python
#now to concatenate
#bring all strings to literals
#and the remove spaces from 2 litreal strings ex names "sam allen" to "samallen"

#column overview to split
data['overview']=data['overview'].apply(lambda x:x.split())

#remove spaces
a=lambda x:[i.replace(" ","") for i in x]
data['genres']=data['genres'].apply(a)
data['keywords']=data['keywords'].apply(a)
data['cast']=data['cast'].apply(a)
data['crew']=data['crew'].apply(a)


#contanate all in one column
data['tags']=data['overview']+data['genres']+data['keywords']+data['cast']+data['crew']
```

```python
data.drop(['overview','genres','keywords','cast','crew'],axis=1,inplace=True)
data
```

```python

data['tags']=data['tags'].apply(lambda x:" ".join(x))
data['tags']=data['tags'].apply(lambda x:x.lower())
```

```python
data
```

# NOW WE DO **VECTORIZATION**

```python
#now we apply stemming to remove repeated words or diffrent forms of same words
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
data['tags']=data['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#convert into a array of words
vectors=cv.fit_transform(data['tags']).toarray()
vectors
```

```python
#now we calculate cosine distance between each vector(movie),
#cosine distance is angle, the bigger the distance the worst the similarity

from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[0] #this has distance of 1st movie with every other movie
```

```python
def recommend(movie):
  movie_index=data[data['title']==movie].index[0]  #getting the index of the given movie to be searched
  distances=similarity[movie_index]       #getting the similarity row of that movie from similarity matrix
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]    #sort those distances in descending order(reverse=t) and get the closest 5 movies
  return movies_list
```
